```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
x_train
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
print(model.predict([[34,10,34]]))
```

```python
# print("slope of tv vs sales:",model.coef_[0])
# print("slope of radio vs sales:",model.coef_[1])
# print("slope of newspeper vs sales:",model.coef_[2])

print("slope is",list((zip(x,model.coef_))))
```

```python
print("constant is",model.intercept_)
```

```python
df1=pd.DataFrame({"actualvalue":y_test,"predicted value":y_pred,"difference":y_test-y_pred})
df1
```

```python
from sklearn.metrics import mean_absolute_error
MAE=mean_absolute_error(y_test,y_pred)
MAE
```

```python
from sklearn.metrics import mean_absolute_percentage_error
MAPE=mean_absolute_percentage_error(y_test,y_pred)
MAPE
```

```python
from sklearn.metrics import mean_squared_error
MSE=mean_squared_error(y_test,y_pred)
MSE
```

```python
from sklearn.metrics import root_mean_squared_error
RMSE=root_mean_squared_error(y_test,y_pred)
RMSE
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_pred,y_test)
score
```
