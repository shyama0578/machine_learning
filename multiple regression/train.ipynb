```python
#get dummies (encoding method inside pandas )
df1=pd.get_dummies(df[['Location','Fuel_Type',"Transmission",'Owner_Type','BRAND']],drop_first=True,dtype=int)#dtype=int 0 and 1
df1
```

```python
df2=pd.concat([df,df1],axis=1)#combine one hot encoding and old col
df2
```

```python
df2.drop(["Name","Location","Unnamed: 0","Fuel_Type","Transmission","Owner_Type","New_Price","BRAND"],inplace=True,axis=1)
df2
```

```python
df2['Mileage']=df2['Mileage'].str.replace('km/kg','')
df2["Mileage"]=df2["Mileage"].str.replace('kmpl',"")
df2['Engine']=df2['Engine'].str.replace("CC",'')
df2['Power']=df2['Power'].str.replace("bhp",'')
df2
```

```python
#null covert to zero(pass object (zero==>False)
df2["Mileage"]=df2['Mileage'].str.replace('null',"0")#dtype object
df2["Power"]=df2['Power'].str.replace("null","0")
df2["Engine"]=df2["Engine"].str.replace("null","0")
df2
```

```python
df2.dtypes
```

```python
#how to convert datatype(object to int)
df2["Mileage"]=df2["Mileage"].astype(float)
df2["Engine"]=df2["Engine"].astype(float)
df2["Power"]=df2["Power"].astype(float)
```

```python
df2.dtypes
```

```python
#null can't directly converted to missing value
#null ==>zero(object)==>float(int)==>missing (here engine power milage can't have zero)
df2.loc[df2.Mileage==0,'Mileage']=np.nan
df2.loc[df2.Engine==0,"Engine"]=np.nan
df2.loc[df2.Power==0,"Power"]=np.nan
df2
```
