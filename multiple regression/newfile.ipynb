```python
y=df.iloc[:,-1]
y
```

```python
# from sklearn.preprocessing import LabelEncoder
# lab=LabelEncoder()
# df["State"]=lab.fit_transform(df['State'])
# df
#here we use onehot encoding
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
col_trans=make_column_transformer((OneHotEncoder(handle_unknown='ignore'),['State']),remainder='passthrough')
x=col_trans.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
x_train
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
df1=pd.DataFrame({"actual value":y_test,"predicted":y_pred,"difference":y_test-y_pred})
df1
```

```python
x=["R&D Spend","Administration","MarketingSpend"]
print("slope is",list((zip(x,model.coef_))))
```

```python
print("constant is",model.intercept_)
```

```python
from sklearn.metrics import mean_absolute_error
MAE=mean_absolute_error(y_pred,y_test)
MAE
```

```python
from sklearn.metrics import mean_absolute_percentage_error
MAPE=mean_absolute_percentage_error(y_pred,y_test)
MAPE
```

```python
from sklearn.metrics import mean_squared_error
MSE=mean_squared_error(y_pred,y_test)
MSE
```

```python
from sklearn.metrics import root_mean_squared_error
RMSE=root_mean_squared_error(y_pred,y_test)
RMSE
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_pred,y_test)
score
```
