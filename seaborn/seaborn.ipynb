```python
#feature selection:select necessery col
correlation=df.corr(numeric_only=True)
correlation #diagonal elemnts are one
```

```python
#correlation matrix graphical resentation ==>heatmap
sns.heatmap(correlation)#cmap="coolwarm"
#light color=>strong correlation
#dark color=weak correlatiom
```

```python
#joinplot==>numerical value
sns.jointplot(x="total_bill",y="tip",data=df,kind="reg")#hex
#combination of hist and scatterplot
#right side tip hist
#top histogram of total_bill
#kind="reg"==>regression plot and distribution pattern
#kind="hex"==>hexagon

```

```python
#pairplot(numerical variable relation multiple)
sns.pairplot(df,hue="sex")#hue ==>specific l data
```

```python
#countplot ==>count of catogorical colm (instead of value count )
sns.countplot(x='sex',data=df,color="k")#only for catogorical data
```

```python
sns.countplot(x='smoker',data=df,color="k")
```

```python
sns.countplot(x='day',data=df,hue="sex")
```

```python
sns.countplot(x='day',data=df,hue="smoker")
```

```python
#barplot==>comparison btwen numerical catog
sns.barplot(x="total_bill",y="day",data=df,ci=None) #confidence interval--ci
#line==>errorbar(if variation increases error bar increases)
```

```python
sns.barplot(x="tip",y="day",data=df)
```

```python
#to check outliers present in the graph we use boxplot

sns.boxplot(x="smoker",y="total_bill",data=df,color="m")
#box==>inter quartile range
#viskers==>line extend from iqr,
#upper viskers(maximum value) and lower viskers(minimum value)
#outliers==>wrondata
#q1=25th percentile
#q2 is medium
#Q3==>75 th percentile
```
